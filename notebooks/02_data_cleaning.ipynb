# Data Cleaning

# Online Retail Analytics

### Objective

In this notebook, we will clean the dataset by handling:

- Missing values
- Duplicates
- Incorrect data types
- Negative quantities
- Invalid prices
- Cancelled orders

This will prepare the dataset for analysis and modeling.

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/raw/online_retail_II.csv")

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna(subset=["Customer ID", "Description"])

df_clean.isnull().sum()
```

```python
df_clean = df_clean.drop_duplicates()

print(df_clean.duplicated().sum())
```

```python
df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"])
```

```python
df_clean = df_clean[df_clean["Quantity"] > 0]
```

```python
df_clean = df_clean[df_clean["Price"] > 0]
```

```python
df_clean["TotalPrice"] = df_clean["Quantity"] * df_clean["Price"]
```

```python
print(df_clean.shape)
df_clean.head()
```

```python
output_path = Path("../data/processed/cleaned_online_retail.csv")

df_clean.to_csv(output_path, index=False)
```
