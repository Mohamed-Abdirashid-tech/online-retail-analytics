# Exploratory Data Analysis

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/processed/cleaned_online_retail.csv")

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
```

```python
df["Country"].value_counts().head(10)
```

```python
df["Description"].value_counts().head(10)
```

```python
total_revenue = df["TotalPrice"].sum()
```

```python
from IPython.display import display

display(f"💰 Total Revenue: ${total_revenue:,.2f}")
```

### Insight

The dataset generated a total revenue of **$17,374,804.27** during the observed period.
This metric serves as the primary KPI for evaluating overall business performance.

```python
df.groupby("Customer ID")["TotalPrice"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10)
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Month"] = df["InvoiceDate"].dt.month
```

```python
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Day"] = df["InvoiceDate"].dt.day
df["Hour"] = df["InvoiceDate"].dt.hour
df["Weekday"] = df["InvoiceDate"].dt.day_name()
```

```python
df.head(10)
```

```python
monthly_sales = df.groupby("Month")["TotalPrice"].sum()

monthly_sales
```

```python
monthly_sales.describe()
```

```python
average_order_value = (
    df.groupby("Invoice")["TotalPrice"]
      .sum()
      .mean()
)

print(f"Average Order Value: ${average_order_value:,.2f}")
```

### Insight

Average Order Value (AOV) measures the average amount customers spend per order.

A higher AOV generally indicates stronger purchasing behavior.

```python
total_customers = df["Customer ID"].nunique()

print(f"Total Customers: {total_customers:,}")
```

```python
top_customers = (
    df.groupby("Customer ID")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers
```

```python
customer_orders = (
    df.groupby("Customer ID")["Invoice"]
      .nunique()
      .sort_values(ascending=False)
      .head(10)
)

customer_orders
```

### Insight

A small group of customers contributes a significant portion of total revenue,
making them high-value customers for retention strategies.

```python
unique_products = df["Description"].nunique()

print(f"Unique Products: {unique_products:,}")
```

```python
top_products = (
    df.groupby("Description")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
top_quantity = (
    df.groupby("Description")["Quantity"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_quantity
```

```python
countries = df["Country"].nunique()

print(f"Countries: {countries}")
```

```python
country_revenue = (
    df.groupby("Country")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
)

country_revenue.head(10)
```

```python
country_orders = (
    df.groupby("Country")["Invoice"]
      .nunique()
      .sort_values(ascending=False)
)

country_orders.head(10)
```

```python
monthly_revenue = (
    df.groupby("Month")["TotalPrice"]
      .sum()
)

monthly_revenue
```
